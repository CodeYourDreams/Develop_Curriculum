### Matplotlib

Matplolib is Python 2D plotting library. To make statistical inferences, it extremely crucial to visualize the data.
It works with Python and NumPy. The most used module of Matplotlib is Pyplot. 
https://matplotlib.org/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py

```python
#Make a simple plot
#We are using NumPy library as our data is in the form of arrays.
#plot() method takes three input arguments. 
#We pass two arrays as input arguments in Pyplot's plot() method
#The %matplotlib is a jupyter notebook specific command that let's you
#see the plots in the notebook itself.

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.plot([1,2,3,4,5,6,10])
plt.show()
```

```python
#To add title, label fo x and y axis
plt.plot([1,2,3,4,5,6,10])
plt.title("Line Plot")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()

```

#Part of Matplotlib plot
#Figure : The whole figure which keeps track of all the axes(plots). A figure can have any number of axes.
#Axes : a plot. It is the space in the figure where the data is plotted.
#Axis : There are the number-line-like objects and take care of generating
the graph limits. 


```python
#To make scatterplot
#Here, we will use the third input argument. This argument is for color,
#marker and line. 
plt.plot([1,2,3,4,5], [1,2,3,4,10],'go')
plt.show()
```

```python
# Draw two scatterplots

plt.plot([1,2,3,4,5], [1,2,3,4,10], 'go', label='GreenDots')  
plt.plot([1,2,3,4,5], [2,3,4,5,11], 'b*',label='Bluestars')  
plt.title('A Scatterplot With Legend')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend() 
plt.show()
```

```python
#To increase the size of figure using figure() method
plt.figure(figsize=(12,8)) # width = 12, height = 8
plt.plot([1,2,3,4,5], [1,2,3,4,10], 'go', label='GreenDots')  
plt.plot([1,2,3,4,5], [2,3,4,5,11], 'b*',label='Bluestars')  
plt.title('A Scatterplot With Legend')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend() 
plt.show()
```

```python
#To draw 2 plots in different panels using plt.subplot(1,2)
#This will draw two subplots inside the figure

#Create Figure and Subplots
#subplot() takes 3 arguments - nrows, ncols, index

plt.subplot(1,2,1)
plt.plot([1,2,3,4],[1,4,9,16],"go")
plt.title("Subplot 1")

plt.subplot(1,2,2)
plt.plot([1,2,3,4],[1,4,9,16],"go")
plt.title("Subplot 2")

plt.show()

```

```python
#Another way of creating subplots is by using subplots()
#It takes two arguments nrows and ncols

#The following command will create two empty figures
fig, (ax1,ax2) = plt.subplots(1,2,figsize = (10,4))

```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize = (10,4))

ax1.plot([1,2,3,4,5], [1,2,3,4,10], 'go')  
ax2.plot([1,2,3,4,5], [2,3,4,5,11], 'b*')  

ax1.set_title('Scatterplot 1 ')
ax2.set_title('Scatterplot 2 ')
ax1.set_xlabel('X')
ax2.set_xlabel('X')  
ax1.set_ylabel('Y');  ax2.set_ylabel('Y')  
plt.show()
```

```python
#To show the distribution of the categorical data using bar graph

fruits = (['apple','orange','bananas','strawberry','grapes'])
num_sold = [25,15,30,19,15]

plt.bar(fruits, num_sold)
plt.title("Bar Graph")
plt.xlabel("Fruits")
plt.ylabel("Number of Sold")
plt.show()
```

```python
#To make pie chart using (pie). 
#Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0) 

plt.pie(sizes, explode = explode, labels = labels )
plt.show()
```

```python
#Histogram is an excellent tool for visualizing and understanding 
#the distribution of the data spread
x  = np.random.randn(1000)

plt.hist(x,10)
plt.title("Histogram")
plt.xlabel("Data")
plt.ylabel("Frequency")
plt.show()
```
